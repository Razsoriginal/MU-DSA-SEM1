```python
#Transpose function-Converts rows into columns and vice versa
import pandas as pd
import numpy as np
# creating a dataframe
data = {
    'Name' : pd.Series(['Swapnil', 'Sameer', 'Omkar', 'Aditya']),
    'Language' : pd.Series(['English', 'Hindi', 'Marathi', 'French']),
    'Marks' : pd.Series([87,88,89, 90])
}
#create a dataframe
df = pd.DataFrame(data)
print(df)
print(df.T)
```

```python
print(df)
```

```python
# say i want to fetch only the marks 
# variable from my dataframe
Marks = df["Marks"]
print(Marks)
```

```python
# selecting multiple coloumns from a dataframe
subset = df[["Name","Language"]]
print(subset)
```

```python
# selecting specific row and creating a subset of them
more_than_88 = df[df["Marks"]>88]
print(more_than_88)
```
