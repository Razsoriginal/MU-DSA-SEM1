```python
import pandas as pd
import numpy as np
df = pd.read_csv('C:/Users/swapn/OneDrive/Desktop/MSC[CC]/Sem I/DSA/Dataset/automobile.csv')
df.head()
```

```python
from sklearn.linear_model import LinearRegression

# Fit model on original df
lm = LinearRegression()
x = df[['highway-mpg']]
y = df['price']
lm.fit(x, y)

# Fit model on df
lm_for_copy = LinearRegression()
x1 = df[['highway-mpg']]
y1 = df['price']
lm_for_copy.fit(x1, y1)

# Predict on original data (optional)
lm_for_copy.predict(x)

# Extract intercept and slope
intercept = lm_for_copy.intercept_
slope = lm_for_copy.coef_[0]  # coef_ is an array, extract value
print(intercept)
print(slope)

# Predict price for user input
desired_mpg = int(input('Enter a highway_mpg for predicting price: '))
predicted_price = slope * desired_mpg + intercept
print(predicted_price)
```

```python
# Multiple Regression
multi_reg_model = LinearRegression()

X = df[['highway-mpg', 'horsepower']]
Y = df['price']

multi_reg_model.fit(X,Y)
multi_reg_model.predict(X)


intercept = multi_reg_model.intercept_
slope = multi_reg_model.coef_
print(intercept)
print(slope)

desired_mpg = int(input('Enter a highway_mpg for predicting price: '))
horsepower = int(input('Enter a horsepower for predicting price: '))
y = (slope[0] * desired_mpg + intercept) + (slope[1] * horsepower + intercept)

print(y)
```
