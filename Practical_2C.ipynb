```python
#plotting with matplotlibs pylot
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Markers
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([13,18,11,20])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# Line type
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([13, 18, 11, 20])
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
#  Matplotlib Labels and Title
import matplotlib.pyplot as plt
import numpy as np

x = np.array([70,75,80,90, 95, 100, 105, 110, 115, 120])
y = np.array([220, 230, 240, 250, 260, 270, 280, 290,300,310])

plt.plot(x, y)
plt.title("Sports Watch Data")
plt.xlabel("Average Pluse")
plt.ylabel("Calorie Burnage")

# plt.show()

# Add Grid Line to a Plot
plt.grid()
plt.show()
```

```python
# Add Grid Line to a Plot
plt.grid()
plt.show()
```

```python
# Mathplotlib Subplot
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 2, 5, 6])
y = np.array([2,6,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

#plot 2:
x= np.array([0, 4, 5, 6])
y = np.array([10, 20, 30, 40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
import numpy as np

x = np.array([15,17,18,17,12,17,12,19,14,20,23,19,16])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()

```

```python
# Bar Plot
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([12, 17, 11, 20])

plt.bar(x,y)
plt.show()

```

```python
# Histogram Plot
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(120, 30, 280)

plt.hist(x)
plt.show() 

```

```python
# Pie Chart 
import matplotlib.pyplot as plt
import numpy as np

y = np.array([25, 50, 125, 75])

plt.pie(y)
plt.show() 

```
